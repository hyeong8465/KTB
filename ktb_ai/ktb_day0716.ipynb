```python
'''
머신러닝3
- 비지도학습
데이터 간의 유사성을 찾고 데이터를 군집, 압축하는 등의 작업 수행
클러스터링, 차원축소, 연관 규칙 학습
장점: 데이터 탐색 및 새로운 패턴 발견에 유용하다.
단점: 결과 해석, 성능 평가, 정확성 보장이 어렵다.(주관적 해석이 필요)

- k-means clustering
특징: 군집 수 K 직접 설정, 거리 기반 알고리즘, 평균 중심, 반복적 과정, 계산 효율성
기본원리: 초기화, 할당, 업데이트, 반복
목적함수: 각 군집의 총 거리의 제곱합을 최소화하는 방향으로 작동
장점: 간단하고 빠름, 해석 용이, 확장성
단점: K 결정, 초기 중심에 민감, 구형 군집, 노이즈와 이상치에 민감
-> K-Means++ 초기화(초기 군집), 엘보우 방법(K 결정), 실루엣 분석(군집화 정도)


- 계층적 군집 분석
데이터 포인트 간의 유사도를 바탕으로 계층적 군집 구조를 형성하는 군집화 방법
병합적 군집화 bottom up, 분할적 군집화 top down
군집 간의 거리측정: 단일 연결, 완전 연결, 평균 연결, 중심 연결, 워드 연결
계층적 군집 분석 기본원리: 초기화 -> 거리 계산 -> 군집 병합(거리 기준)
장점: 계층적 구조 시각화 가능, 군집 수 결정 불필요, 유연성
단점: 계산 비용, 병합 후 수정 불가, 노이즈에 민감

- DBSCAN(Density-Based Spaial Clustering of Application with Noise)
밀도 기반의 군집화 알고리즘, 밀집된 영역을 군집으로 식별하기 때문에 군집의 크기나 형태에 구애받지 않는 유연한 군집화
특징: 밀도 기반 군집화, 노이즈 처리, 유연한 군집 형성, 간단한 구현
주요 하이퍼파라미터: eps(데이터 포인트 간의 최대 거리), minpts(군집을 형성하기 위해 필요한 최소 데이터 포인트 수)
장점: 군집 수 미리 지정 불필요, 노이즈 데이터 처리, 다양한 군집 형태, 확장성
단점: 하이퍼파라미터 설정 어려움, 밀도 변화에 민감, 고차원 데이터에서 성능 저하

- PCA
고차원 데이터를 저차원으로 변환하여 데이터의 주요 변동성을 보존하는 차원 축소 기법
분산을 최대화하는 직교 축을 찾는다.
기본원리: 데이터 중심화, 공분산 행렬, 고유값 분해, 주성분 선택, 데이터 변환
장점: 차원 축소, 노이즈 제거, 데이터 압축
단점: 해석의 어려움, 선형성 가정, 정보 손실
-> 비선형 차원 축소 기법, 고차원 데이터 처리, PCA변형(커널 PCA, Sparse PCA)

- Isomap
비선형 차원 축소 기법
기본원리: 최근접 이웃 그래프 구성, 지오데식 거리 계산(경로 계산), 다차원 척도법
장점: 비선형 구조 보존, 지오데식 거리사용, 고차원 데이터 시각화
단점: 계산 복잡성, 노이즈에 민감, k값 결정
-> 최적의 k값 선택, 노이즈 제거, 고차원 데이터 처리

- t-SNE(t-Distributed Stochastic Neigbor Embedding)
데이터 포인트 간의 지역적 유사성을 보존하는 데 강력한 성능
기본원리: 고차원 공간에서 거리 계산 및 확률 변환, 저차원 공간에서 거리 계산 및 확률 변환, 확률 분포 간의 차이 최소화
장점: 고차원 데이터 시각화, 지역적 유사성 보존, 다양한 데이터 유형 처리
단점: 시간 복잡도, 매개변수 민감성, 해석의 어려움

- 연관 규칙 학습(Association Rule learning)
특정 항목이 나타날 때 다른 항목이 함께 나타날 확률을 계산하여 유용한 패턴을 도출
Apriori 알고리즘, FP-Growth 알고리즘
특징: 빈발 항목 집합 기반, 계산 효율성, 응용 가능성
기본 원리: 빈발 항목 집합 탐색, 연관 규칙 생성
장점: 유용한 패턴 발견, 다양한 응용 분야, 단순성과 이해 용이성
단점: 계산 복잡성, 희소성 문제, 과적합
-> 알고리즘 최적화, 전처리, 평가 지표 다양화

비지도 학습의 경우 데이터의 패턴을 학습하기 때문에 노이즈에 굉장히 민감하다.
'''
```
